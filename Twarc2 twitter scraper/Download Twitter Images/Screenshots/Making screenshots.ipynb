```python
import pandas as pd
pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_columns', None)
pd.options.mode.chained_assignment = None
```

```python
ht = pd.read_csv('felizmiercoles.txt')
```

```python
ht = ht[(ht['media keys'] != '[]') & (ht['media urls'] == '[]')]
```

```python
ht['tweet urls'] = ht['tweet urls'].str.replace('[','',regex=True)
ht['tweet urls'] = ht['tweet urls'].replace(']','',regex=True)
ht['tweet urls'] = ht['tweet urls'].replace('\'','',regex=True)
ht['tweet urls'] = ht['tweet urls'].str.split(',')
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from time import sleep
options = Options()
options.add_argument(r'--user-data-dir=C:\Users\Piratena\Desktop\Python laboratory\UserData')
options.page_load_strategy = 'normal'
driver = webdriver.Chrome(options=options)
driver.get("https://www.twitter.com/")
sleep(30)
```

```python
def get_screenshot(row):
    author = row['author']
    urls = row['tweet urls']
    id = row['id tweet']
    for url in urls:
        driver.get(url)
        sleep(3)
        driver.get_screenshot_as_file(f'screenshot_{author}_{id}.png')
```

```python
def remove_url(row):
    urls = row['tweet urls']
    while len(urls) > 1:
        urls.remove(urls[-1])
```

```python
ht = ht[:10]
```

```python
ht
```

```python
ht.apply(remove_url,axis=1)
```

```python
ht.apply(get_screenshot,axis=1)
```
