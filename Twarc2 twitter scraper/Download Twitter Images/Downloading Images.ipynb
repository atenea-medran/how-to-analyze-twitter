# Descargar imagenes autom√°ticamente

```python
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns
import re
import unidecode as ud
import requests

pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_columns', None)
pd.options.mode.chained_assignment = None
```

Guardamos el hashtag en una variable.

```python
hashtag = 'warzone2'
```

Abrimos el archivo .txt con pandas para tener los datos en formato tabla (DataFrame).

```python
ht = pd.read_csv(f'{hashtag}.txt')
```

```python
ht
```

```python
def descargar_avatar(row):
    author = row['author']
    avatar_url = row['profile image url']
    avatar_url = avatar_url.replace('normal','400x400')
    avatar_data = requests.get(avatar_url).content
    with open(f'Avatares/avatar_{author}.jpg','wb') as handler:
        handler.write(avatar_data)
```

```python
ht.apply(descargar_avatar,axis=1)
```

```python
ht_media = ht.copy()
ht_media = ht_media[ht_media['media urls'] != '[]']
ht_media
```

```python
ht_media['media urls'] = ht_media['media urls'].str.replace('[','',regex=False).str.replace(']','',regex=False).str.replace('\'','',regex=False)
ht_media['media urls'] = ht_media['media urls'].str.split(',')
```

```python
def descargar_imagen(row):
    author = row['author']
    urls = row['media urls']
    for url in urls:
        image_data = requests.get(url).content
        with open(f'Imagenes/{author}.jpg','wb') as handler:
            handler.write(image_data)
```

```python
ht_media.apply(descargar_imagen,axis=1)
None
```
