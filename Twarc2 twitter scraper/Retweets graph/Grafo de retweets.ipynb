# Grafo de retweets

```python
import pandas as pd
import glob
import unidecode as ud
import re
import json

pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_columns', None)
pd.options.mode.chained_assignment = None
```

Guardamos en una variable el nombre de nuestro hashtag o palabra clave.

```python
hashtag = 'felizmiercoles'
```

```python
tweets_data = []

tweets_file = open(f'../{hashtag}.jsonl', encoding = 'utf-8')

for line in tweets_file:
    tweet = json.loads(line)
    tweets_data.append(tweet)

tweets_file.close()
```

```python
list_users_flattered = []

for item in tweets_data:
    list_users = item['includes']['users']
    for user in list_users:
        if 'location' in user:
            location = user['location']
        else:
            location = 'None'
        username = user['username']
        verified = user['verified']
        name = user['name']
        id_user = user['id']
        profile_image_url = user['profile_image_url']
        followers_count = user['public_metrics']['followers_count']
        following_count = user['public_metrics']['following_count']
        tweet_count = user['public_metrics']['tweet_count']
        listed_count = user['public_metrics']['listed_count']
        bio = user['description']
        created_at = user['created_at']
        if 'url' in user:
            url_list = user['entities']['url']['urls']
            url_end = []
            for url_unique in url_list:
                if 'expanded url' in url_unique:
                    url_end.append(url_unique['expanded_url'])
                elif 'expanded url' not in url_unique:
                    url_end.append(url_unique['url'])
        else:
            url_end = 'None'

        if 'entities' in user:
            if 'description' in user['entities']:
                if 'urls' in user['entities']['description']:
                    url_list2 = user['entities']['description']['urls']
                    url_end2 = []
                    for url_unique2 in url_list2:
                        url_end2.append(url_unique2['expanded_url'])
                else:
                    url_end2 = 'None'
        else:
            url_end2 = 'None'
        list_users_flattered.append([location,username,verified,name,id_user,profile_image_url,
        followers_count,following_count,tweet_count,listed_count,bio,created_at,url_end,url_end2])
```

```python
users_df = pd.DataFrame(list_users_flattered,columns=['location','author','verified','name',
'id author','profile image url','followers','following','tweets','listed',
'description','created at','author url','description urls'])
users_df = users_df.drop_duplicates(subset='id author')
```

Abrimos el archivo del hashtag en pandas.

```python
ht = pd.read_csv(f'../{hashtag}.txt',on_bad_lines='skip')
print('Número de tweets:',len(ht))
ht.head(2)
```

```python
info_users = users_df.drop_duplicates(subset='author',keep='first')
info_users = info_users[['author','name','id author','created at',
                         'author url','description urls','profile image url','followers','following','tweets']]
#info_users['name'] = info_users['name'].apply(lambda x:ud.unidecode(x))
info_users['id author'] = info_users['id author'].apply(lambda x:str(x))
```

```python
info_users
```

```python
ht = ht[ht['relation'] == 'retweeted']
```

```python
links_in = ht['user retweeted'].value_counts().to_frame().reset_index()
links_in.columns = ['author','links in']
links_in
```

```python
nodos_0_links_in = ht['author'].to_frame().drop_duplicates()
links_in = pd.concat([links_in,nodos_0_links_in],axis=0)
links_in = links_in.drop_duplicates(subset='author',keep='first')
```

```python
links_out = ht['author'].value_counts().to_frame().reset_index()
links_out.columns = ['author','links out']
links_out
```

```python
nodos = links_in.merge(links_out,on='author',how='left')
```

```python
nodos['links in'] = nodos['links in'].fillna(0)
nodos['links out'] = nodos['links out'].fillna(0)
```

```python
nodos['total links'] = nodos['links in'] + nodos['links out']
nodos = nodos.reset_index()
nodos = nodos[['index','author','total links','links in','links out']]
nodos['total links'] = nodos['total links'].apply(lambda x:int(x))
nodos['links in'] = nodos['links in'].apply(lambda x:int(x))
nodos['links out'] = nodos['links out'].apply(lambda x:int(x))
```

```python
nodos = nodos.merge(info_users,on='author',how='left')
```

```python
nodos
```

```python
nodos['name'] = nodos['name'].fillna('None')
nodos['id author'] = nodos['id author'].fillna('None')
nodos['created at'] = nodos['created at'].fillna('None')
nodos['author url'] = nodos['author url'].fillna('None')
nodos['description urls'] = nodos['description urls'].fillna('None')
nodos['profile image url'] = nodos['profile image url'].fillna('None')
nodos['followers'] = nodos['followers'].fillna(0)
nodos['following'] = nodos['following'].fillna(0)
nodos['tweets'] = nodos['tweets'].fillna(0)

nodos['followers'] = nodos['followers'].apply(lambda x:int(x))
nodos['following'] = nodos['following'].apply(lambda x:int(x))
nodos['tweets'] = nodos['tweets'].apply(lambda x:int(x))
```

```python
titulo_nodos = 'nodedef>name VARCHAR,label VARCHAR,total links INT,links in INT,links out INT,name VARCHAR,id user VARCHAR,created at VARCHAR,user url VARCHAR,description urls VARCHAR,avatar VARCHAR,followers INT,following INT,tweets INT\n'
```

```python
with open(f'grafo_retweets_{hashtag}.gdf','w') as file:
    file.write(titulo_nodos)
```
