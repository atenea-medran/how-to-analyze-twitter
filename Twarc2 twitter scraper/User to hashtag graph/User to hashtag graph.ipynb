# Grafo de usuarios a hashtags

```python
import pandas as pd
import glob
import unidecode as ud
import re
import json

pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_columns', None)
pd.options.mode.chained_assignment = None
```

Guardamos en una variable el nombre de nuestro hashtag o palabra clave.

```python
hashtag = 'felizmiercoles'
```

```python
tweets_data = []

tweets_file = open(f'../{hashtag}.jsonl', encoding = 'utf-8')

for line in tweets_file:
    tweet = json.loads(line)
    tweets_data.append(tweet)

tweets_file.close()
```

```python
list_users_flattered = []

for item in tweets_data:
    list_users = item['includes']['users']
    for user in list_users:
        if 'location' in user:
            location = user['location']
        else:
            location = 'None'
        username = user['username']
        verified = user['verified']
        name = user['name']
        id_user = user['id']
        profile_image_url = user['profile_image_url']
        followers_count = user['public_metrics']['followers_count']
        following_count = user['public_metrics']['following_count']
        tweet_count = user['public_metrics']['tweet_count']
        listed_count = user['public_metrics']['listed_count']
        bio = user['description']
        created_at = user['created_at']
        if 'url' in user:
            url_list = user['entities']['url']['urls']
            url_end = []
            for url_unique in url_list:
                if 'expanded url' in url_unique:
                    url_end.append(url_unique['expanded_url'])
                elif 'expanded url' not in url_unique:
                    url_end.append(url_unique['url'])
        else:
            url_end = 'None'

        if 'entities' in user:
            if 'description' in user['entities']:
                if 'urls' in user['entities']['description']:
                    url_list2 = user['entities']['description']['urls']
                    url_end2 = []
                    for url_unique2 in url_list2:
                        url_end2.append(url_unique2['expanded_url'])
                else:
                    url_end2 = 'None'
        else:
            url_end2 = 'None'
        list_users_flattered.append([location,username,verified,name,id_user,profile_image_url,
        followers_count,following_count,tweet_count,listed_count,bio,created_at,url_end,url_end2])
```

```python
users_df = pd.DataFrame(list_users_flattered,columns=['location','author','verified','name',
'id author','profile image url','followers','following','tweets','listed',
'description','created at','author url','description urls'])
users_df = users_df.drop_duplicates(subset='id author')
```

```python
ht = pd.read_csv(f'../{hashtag}.txt',on_bad_lines='skip')
print('Número de tweets:',len(ht))
ht.head(2)
```

```python
info_users = ht.drop_duplicates(subset='author',keep='first')
info_users = info_users[['author','name','id author','description','location','created at',
                         'author url','description urls','profile image url','followers','following','tweets']]
info_users['id author'] = info_users['id author'].apply(lambda x:str(x))
```

```python
ht
```

```python
ht['text'] = ht['text'].fillna('None')
ht['hashtags'] = ht['text'].apply(lambda x: re.findall(r"#(\w+)", x))
hashtag_list = ht['hashtags'].tolist()
ht_con_tildes = []
for listed in hashtag_list:
    for ht2 in listed:
        ht_con_tildes.append(ht2)
hashtag_list_separated = []
for ht2 in ht_con_tildes:
    ht2 = ud.unidecode(ht2)
    hashtag_list_separated.append(ht2)
```

```python
hashtags = pd.DataFrame (hashtag_list_separated, columns = ['label']) 
hashtags['label'] = hashtags['label'].str.lower()
hashtags
```

```python
hashtags = hashtags.value_counts().to_frame().reset_index()
hashtags = hashtags.rename(columns={0:'count'})
hashtags['label'] = hashtags['label'].apply(lambda x:'#'+x)
hashtags
```

```python
users = ht['author'].to_frame().drop_duplicates()
users.columns = ['label']
users['count'] = 0
users['label'] = users['label'].apply(lambda x:'@'+x)
users
```

```python
nodos = pd.concat([hashtags,users])
nodos = nodos.reset_index().drop('index',axis=1).reset_index()
nodos
```

```python
titulo_nodos = 'nodedef>name VARCHAR,label VARCHAR,links in INT\n'
```

```python
with open(f'grafo_user_to_hashtag_{hashtag}.gdf','w') as file:
    file.write(titulo_nodos)
```

```python
nodos.to_csv(f'grafo_user_to_hashtag_{hashtag}.gdf',sep=',',index=False,header=False,mode='a')
```

```python
def construir_ejes(row):
    source = '@' + row['author']
    target = row['hashtags']
    directed = 'True'
    for ht in target:
        ht = ud.unidecode(ht)
        ht = '#' + ht.lower()
        ejes_list.append([source,ht,directed])
```

```python
ejes_list = []

ht.apply(construir_ejes,axis=1)
None
```

```python
ejes = pd.DataFrame(ejes_list,columns=['source','target','directed'])
ejes
```

```python
index.columns = ['source','source index']
```

```python
ejes = ejes.merge(index,on='source',how='left')
```

```python
index.columns = ['target','target index']
```

```python
ejes = ejes.merge(index,on='target',how='left')
```

```python
ejes.columns = ['source label','target label','directed','source','target']
ejes = ejes[['source','target','source label','target label','directed']]
ejes
```

```python
ejes['weight'] = ejes.groupby(['source','target']).source.transform('size')
ejes['source']
ejes
```

```python
ejes = ejes.drop_duplicates(subset=['source','target'])
```

```python
ejes = ejes.sort_values(by='weight',ascending=False)
```

```python
titulo_ejes = 'edgedef>node1 VARCHAR,node2 VARCHAR,node1 name VARCHAR,node2 name VARCHAR,directed BOOLEAN,weight INT\n'
```

```python
with open(f'grafo_user_to_hashtag_{hashtag}.gdf',mode='a') as file:
    file.write(titulo_ejes)
```
