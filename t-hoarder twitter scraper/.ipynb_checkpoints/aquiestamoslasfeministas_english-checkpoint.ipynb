## Vista general hashtag

### Cosas que no están automatizadas: poner archivo clean_ok.txt y archivo modularidad.csv

Abrimos el archivo .txt con pandas para tenerla en formato tabla (DataFrame).

```python
ht = pd.read_csv('aquiestamoslasfeministas_clean_ok.txt', sep="\t", on_bad_lines="skip")
```

Con t-hoarder y Gephi hemos obtenido un grafo del que podemos sacar una tabla con información de los usuarios que participan en el hashtag. Abrimos el archivo de modularidad como tabla. Extraemos la columna 'author' y 'modularity_class' para añadir esta última a la tabla principal. Convertimos las columnas 'author', 'text' y 'user retweeted' en minúsculas para evitar errores futuros. Los participantes del hashtag se agrupan por afinidad en clases y esta la podemos conocer gracias a la columna 'modularity class'.

```python
ht_modularity = pd.read_csv('aquiestamoslasfeministas_modularidad.csv')
ht_modularity = ht_modularity[['Label', 'modularity_class']]
ht_modularity.columns=['author', 'modularity_class']
ht_modularity['author'] = ht_modularity['author'].str.lower()
ht['author'] = ht['author'].str.lower()
ht['text'] = ht['text'].str.lower()
ht['user retweeted'] = ht['user retweeted'].str.lower()
ht = ht.merge(ht_modularity, how="left", on="author")
ht.head()
```

A partir de la columna 'date' creamos otra con solo la fecha y la hora, sin minutos ni segundos.

```python
ht['date_calc'] = ht['date'].apply(lambda x: x.split(":")[0])
ht['date_calc'] = ht['date_calc'].apply(lambda x: x + ":00")
```

Con esa nueva columna construimos una tabla dinámica para calcular cuántos tweets de cada clase hay por hora.

```python
tweets_hour = pd.pivot_table(ht,index= "date_calc",columns= "modularity_class",aggfunc= ["count"],values=['id tweet'])
tweets_hour = tweets_hour.fillna(0)
tweets_hour.to_csv('distribucion_por_horas.csv')

tweets_hour = pd.read_csv('distribucion_por_horas.csv')
tweets_hour.columns = tweets_hour.iloc[1]
tweets_hour = tweets_hour[3:]
tweets_hour.to_csv('distribucion_por_horas.csv')

tweets_hour = pd.read_csv('distribucion_por_horas.csv')
tweets_hour = tweets_hour.drop(['Unnamed: 0'], axis=1)
tweets_hour.rename(columns={'modularity_class':'hour'}, inplace=True)
tweets_hour.head()
```

Representamos el número de tweets por hora en un gráfico de barras acumuladas para identificar diferentes picos de actividad. 

```python
ax = tweets_hour.plot.bar(x='hour', stacked=True,figsize=(15,7),legend=None)
ax.spines[['top','right']].set_visible(False)
plt.text(-0.4,1720,'Cronología completa',fontsize=24)
plt.text(-0.55,1640,'El primer pico es el pico TERF que nos interesa.',fontsize=14)
plt.xlabel('Horas',fontsize=14)
plt.tick_params(axis='x',which='both',bottom=False,top=False,labelbottom=False)
plt.ylabel('Número de tweets',fontsize=14)
plt.savefig('1. cronologia completa.jpg',bbox_inches='tight',dpi=200)
```

# Calcular volúmenes de autores

```python
alternativa = [['prueba','0','0','0','0']]
```

```python
alternativa = pd.DataFrame(alternativa,columns=['author','total','original_tweets','rts_made','rts_received'])
alternativa
```

Convertimos los autores en una lista para loopear.

```python
author_lista = ht_terf['author'].unique().tolist()
```

Loopeamos para obtener una lista de listas con el autor, el total de aportaciones, los tweets originales y los retweets recibidos.

```python
authors_list_of_lists = []

for author in author_lista:
    only_author = ht_terf[ht_terf['author'] == author]
    pattern_rt_author = 'rt '+str(author)
    total_len = len(only_author)
    original_tweets = only_author[~only_author['text'].str.startswith("rt @")]
    original_tweets_len = len(original_tweets)
    rts_made_len = total_len - original_tweets_len
    rts_received = ht_terf[ht_terf['text'].str.startswith(pattern_rt_author)]
    rts_received_len = len(rts_received)
    authors_list_of_lists.append([author,total_len,original_tweets_len,rts_made_len,rts_received_len])
```

Convertimos la lista de listas en tabla.

```python
volume = pd.DataFrame(authors_list_of_lists,columns=['author','total','original_tweets','rts_made','rts_received'])
volume
```

Vemos el número de autores.

```python
authors_unique = len(ht_terf['author'].unique())
authors_unique
```

### Tweets originales

Es probable que se copien tweets en vez de hacer retweet y esos no los vamos a contar como tweets originales. Vamos a comprobarlo.

Primero quitamos los retweets.

```python
ht_terf_copypaste = ht_terf[~ht_terf['text'].str.startswith("rt @")]
```

Quitamos los enlaces del final porque hacen a los tweets diferentes para la API de Twitter, pero en la interfaz de usuario son iguales.

```python
ht_terf_copypaste['text'] = ht_terf_copypaste['text'].apply(lambda x:re.sub(r'http\S+', '', x))
```

Descartamos los duplicados manteniendo los primeros, que serán de los autores originales porque están ordenados por fecha.

```python
ht_terf_copypaste = ht_terf_copypaste.drop_duplicates(subset='text',keep='first')
```

Contamos los tweets originales.

```python
true_originals_count = len(ht_terf_copypaste)
true_originals_count
```

Contamos los tweets que parecían originales.

```python
fake_originals_count = volume['original_tweets'].sum()
fake_originals_count
```

Restamos los falsos a los verdaderos.

```python
tweets_copiados = fake_originals_count - true_originals_count
print(tweets_copiados,'twets han sido copiados.')
```

Contamos los tweets originales y sacamos el top 20.

```python
true_original_top20 = ht_terf_copypaste['author'].value_counts().head(20).sort_values(ascending=False)
true_original_top20
```

Contamos los tweets originales del top 20 y calculamos el porcentaje que suponen.

```python
true_originals_top20_count = true_original_top20.sum()
true_originals_top20_count
```

```python
true_originals_top20_percentage = (true_originals_top20_count/true_originals_count*100).round(1)
true_originals_top20_percentage
```

Aparte calculamos la media de tweets originales de todos los autores.

```python
true_originals_mean = round((true_originals_count/authors_unique),1)
true_originals_mean
```

Representamos en un gráfico.

```python
true_original_top20_sorted = true_original_top20.sort_values(ascending=True)

ax = true_original_top20_sorted.plot.barh(x='author',y='total',figsize=(15,7),color='orange',label='',fontsize=14)
ax.spines[['top','bottom','left','right']].set_visible(False)
plt.tick_params(axis='x',which='both',bottom=False,top=True,labelbottom=False,labeltop=True)
plt.tick_params(axis='y',which='both',left=False,right=False)
plt.xlabel('',fontsize=14)
plt.ylabel('')
plt.text(0,22,'Los 20 usuarios con más tweets originales escriben el {} % del total'.format(true_originals_top20_percentage),
         fontsize=23.8)
plt.text(0,21,'Usuarios frente a número de tweets originales.',fontsize=14)
plt.axvline(x=true_originals_mean,color='blue',linewidth=4,label='Media: {} tweets'.format(true_originals_mean))
plt.legend(bbox_to_anchor=(0.99,0.89),prop={'size': 12})
plt.savefig('6.1. originales.jpg',bbox_inches='tight',dpi=100)
```
